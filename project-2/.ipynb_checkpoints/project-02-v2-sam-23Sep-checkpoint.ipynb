### 1.1 Load the file with the `csv` module and put it in a Python dictionary

The dictionary format for data will be the column names as key, and the data under each column as the values.

Toy example:
```python
data = {
    'column1':[0,1,2,3],
    'column2':['a','b','c','d']
    }
```

````python
def load_csv_as_dict(csv) -> dict:
    """Takes in a csv file and returns a dictionary 
       with the table's header as Keys, and table's values as Values"""

    with open(csv) as file:
        raw = file.read()
    
    # Return a list of the csv's table rows
    # Each row/list is a list of 4 strings, corresponding to the table values in each row
    raw_sat_split = [string.split(',') for string in raw.split()]

    header = raw_sat_split[0]
    
    # creates a list of 4 lists, each list is a column in the table
    data = [[row[i] for row in raw_sat_split[1:]] for i in range(len(header))]

    return dict(zip(header, data))

load_csv_as_dict('sat_scores.csv')
````

---

A data dictionary is an object that describes your data. This should contain the name of each variable (column), the type of the variable, your description of what the variable is, and the shape (rows and columns) of the entire dataset.

````python
def data_dictionary(df):
    """ Takes in a DataFrame, and returns the DataFrame's 
        column names, the data type in each column, and description of the data"""

    column_dtypes = (pd.DataFrame(df.dtypes).reset_index()
                                            .rename(columns={'index':'Variables',0:'Type'}))
      
    column_dtypes['Description'] = ['US States', 
                                    '% of students taking SATs', 
                                    'SAT Verbal Avg. Score',
                                    'SAT Math Avg. Score']
    
    print('\nDataFrame has {} rows and {} columns'.format(df.shape[0],df.shape[1]))
    return column_dtypes

data_dictionary(sat)
````

### 6.3 Assign and print the _covariance_ matrix for the dataset

1. Describe how the covariance matrix is different from the correlation matrix.
2. What is the process to convert the covariance into the correlation?
3. Why is the correlation matrix preferred to the covariance matrix for examining relationships in your data?

#### Ans:
1) Both the covariance and correlation matrix measures how one variable varies with another. However, 
- The values in the covariance matrix is relative to the units of the variable, while correlation is a normalization/standardisation of those values.
- Consequently, the value of covariance lies between negative infinity and positive infinitiy, while the value of correlation is between -1 and 1. 
- Covariance is affected by a change in scale, while correlation isn't. 
    
2) Divide $Cov(X,Y)$ by $\sqrt{Var(X)⋅Var(Y)}$ <br>

3) Correlations are a unit-free measure of the relationship between variables, and hence can be used to fairly compare the covariances of multiple variables along the same scale (-1 to 1)


````python
covariance_matrix = sat.drop('State',axis=1).cov()
covariance_matrix
````

#### Summary of data:
- There are no null values
- But there are strings in columns that should be numbers: These are indicated by '-' values in place of empty fields
    - As a result, values in these columns are strings rather than integers/floats. 
    - Need to change these to floats and the '-' values to 0. I'm assuming that no values equate to a median frequency of 0 times
- Age column is a mix of integer values and string values for a range of ages. Problematic because we can't easily compare by ages. For the EDA, I will take these segments as it is, and regroup the age groups if required
    - Unit of analysis is the age segment, rather than the individual

````python
drug = pd.read_csv('drug-use-by-age.csv')
drug
````

Taking a look at the data types for each column, there are certain drug-use/freq statistics that are objects rather than floats

````python
drug.info()
# age is an object, should be integer?
# cocaine-freq, crack-freq, heroin-freq, inhalent-frea etc are objects, shld be floats/ints
````

````python
# find columns/fields where dtype is object
cols_with_objects = [col for col in drug.columns if drug[col].dtype=='object']
cols_with_objects[1:]
````

````python
def to_float(value:str) -> float:
    """
    Converts strings to floats. Where the string doesn't contain a number (ValueError), 
    prints the value and returns 0.
    """
    try:
        return float(value)
    except ValueError:
        print(value)
        return 0
````

````python
for col in cols_with_objects[1:]:
    print(col)
    drug[col] = drug[col].apply(to_float)
````

````python
drug.info()
````

````python
drug.describe()
````

````python
#renaming pain-reliever to get rid of typo
drug.columns = drug.columns.str.replace('pain-releiver','pain_reliever')
drug.columns
````

#### Data Reshaping: to long-format data

````python
def reshape_drug():
    
    # melt drug use/freq columns
    drug_reshape = drug.melt(id_vars=['age','n'])
    
    # separating drug type and statistic type (use/freq) into different columns
    drug_reshape['drug'] = drug_reshape['variable'].apply(lambda x: x.split('-')[0])
    drug_reshape['stat'] = drug_reshape['variable'].apply(lambda x: x.split('-')[1]) 
    drug_reshape.drop(labels='variable',axis=1,inplace=True)
    
    # setting the stat column to wide form -> to see use and freq stat for each drug type
    drug_reshape = drug_reshape.pivot_table(index=['age','n','drug'],columns='stat')
    
    # getting rid of hierarchical columns and renaming column names
    drug_reshape = pd.DataFrame(drug_reshape.to_records())
    drug_reshape.columns = ['age','n','drug','frequency','use']
    
    return drug_reshape
````

````python
drug_reshape = reshape_drug()
drug_reshape.head()
````

````python
def summary_statistics(col):
    """
    Takes in a Series and returns the Mean, Median, Mode, Mode Count and Stddev of the Series
    """
    
    print('Summary Statistics for {}:'.format(col.name))
    print('Mean: {:.2f}'.format(np.mean(col)))
    print('Median: {}'.format(np.median(col)))
    print('Mode: {}'.format(stats.mode(col)[0][0]))
    print('Mode count: {}'.format(stats.mode(col)[1][0]))    
    print('StdDev: {:.2f}'.format(np.std(col)))
````

````python
summary_statistics(drug_reshape['frequency'])
````

**Question:** Are Millennials more likely to consume alcohol than Baby Boomers? <br>

**Deliverables:** 
- % of Millennials and Baby Boomers who have consumed alcohol in the past 12 months
- P-value of Chi-Squared Test used to evaluate if Millennials are significantly more likely to consume alcohol in the past 12 months

**Hypothesis Testing:** <br>
<br>
Note: the drug data is dated 2015, hence the generations will be split by the following age ranges:
- Baby Boomers are 51 - 69 YO (1946 - 1964)
- Millennials are 21 - 34 YO (1981 - 1996, but setting minimum age of 21 for legal drinking age)

````python
# creating a new dataframe that excludes people below 20
drug_hyp = drug_reshape[~drug_reshape['age'].isin([str(x) for x in [12,13,14,15,16,17,18,19,20]])]
````

````python
# new column to categorise Millennials and Baby Boomers
drug_hyp['gen'] = (drug_hyp['age'].apply(lambda x: 'Baby Boomers' 
                                         if x in (['35-49', '50-64', '65+']) else 'Millennials'))
drug_hyp = drug_hyp[['gen','n','drug','use #']]
drug_hyp
````

We will be using a Chi-Squared test to test how independent the rate of use of Alcohol is between Millennials and Baby Boomers, with a Significance Level of Alpha = 0.01

We are employing the Chi_Squared test as we do not know the means of these populations, but we know the categories (age segments)

````python
m_use = drug_hyp[(drug_hyp['drug']=='alcohol') & (drug_hyp['gen']=='Millennials')]['use #'].sum()
m_nouse = drug_hyp[(drug_hyp['drug']=='alcohol') & (drug_hyp['gen']=='Millennials')]['n'].sum() - m_use

b_use = drug_hyp[(drug_hyp['drug']=='alcohol') & (drug_hyp['gen']=='Baby Boomers')]['use #'].sum()
b_nouse = drug_hyp[(drug_hyp['drug']=='alcohol') & (drug_hyp['gen']=='Baby Boomers')]['n'].sum() - b_use
````

````python
m = [m_use, m_nouse]
b = [b_use, b_nouse]
````

````python
print('% of Millennials who use Alcohol: {:.2f}%'.format(m_use/(m_use+m_nouse)*100))
print('% of Baby Boomers who use Alcohol: {:.2f}%'.format(b_use/(b_use+b_nouse)*100))
````

````python
observations = [m,b]
````

````python
import scipy.stats as stats

chi2, p, dof, ex = stats.chi2_contingency(observations, correction=False)
print('p-value: {}'.format(p))
````

````python
print('{:.200f}'.format(p))
````

**Report**

Null Hypothesis: Millennials are not more likely than Baby Boomers to drink Alcohol in the past year <br>

Alternate Hypothesis: Millennials aremore likely than Baby Boomers to drink Alcohol <br>

Significance Level = 0.01<br>

Results: <br>
Based on the p-value of 0, we reject the Null Hypothesis and accept the Alternate hypothesis that there is a statistically significant difference between the % of Millennials who drink Alcohol and Baby Boomers who do.
- Millennials are significantly more more likely to drink alcohol than Baby Boomers at a confidence level of 99%

2. Describe clearly in words the process of calculating the spearman rank correlation.

````python
coef, p = stats.spearmanr(sat['Verbal'],sat['Math'])
````

````python
p
````

1. The spearman correlation determines the strengh and direction of two variables that do not necessarily have to follow a **linear** relationship. The Pearson correlation, on the other hand, measures the strengh of a linear association between two variables. Spearman correlation can also be used for variables that are not normally distributed, unlike Pearson 
2. Process of calculating spearman rank correlation:
    - [Rank the data](https://statistics.laerd.com/statistical-guides/spearmans-rank-order-correlation-statistical-guide.php) for both variables. So for each data point in a variable, the point with the lowest value will be labelled 1 and increases as the value increases. 
    - For tied values, the rank will be the average of the ranks that they would have otherwise occupied
    - Get the difference between ranks, and sum up the square of the differences
    - Lastly calculate the coefficient with the Spearman Rank formula: 1 - ((6 x the sum of squared differences) ÷ n^3/n))
    - https://geographyfieldwork.com/SpearmansRank.htm
