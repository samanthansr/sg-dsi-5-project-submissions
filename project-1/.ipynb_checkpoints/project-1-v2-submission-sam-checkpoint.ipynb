## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def player_power(players_dict, pokedex_dict, player_id, display_stats='No'):
    
    # looking up player's pokemon ids
    player_pokemon_ids = [pokemon_id for pokemon_id in players_dict[player_id]['player_pokemon']]
    
    # initialising dictionary for owned pokemon stats 
    stats = {}
    
    # Finding and aggregating attack, defense and base stat values for each of the player's pokemon(s)
    for pokemon_id in player_pokemon_ids:
        stats[pokemon_id] = {'Attack': 0, 'Defense': 0, 'Base Stats': 0}
        
        if pokemon_id in pokedex_dict.keys():
            stats[pokemon_id]['Attack'] = (pokedex_dict[pokemon_id]['attack'] 
                                           + pokedex_dict[pokemon_id]['special_attack'])
            stats[pokemon_id]['Defense'] = (pokedex_dict[pokemon_id]['defense'] 
                                            + pokedex_dict[pokemon_id]['special_defense'])
            #slice from index 2 to exclude Name and Type
            stats[pokemon_id]['Base Stats'] = sum(list(pokedex_dict[pokemon_id].values())[2:])
        
        else:
            stats[pokemon_id] = 'No stats available in pokedex'
    
    # calulate player power by summing up base statistics of all their pokemon(s)
    player_power = 0
    for pokemon in stats:
        player_power += stats[pokemon]['Base Stats']
    
    print("{}'s power is {}".format(players_dict[player_id]['player_name'],player_power))
    
    # setting optional parameter for users who want to see detailed stats
    if display_stats == 'Yes':
        print(stats)
        return player_power
    else:
        return player_power

player_power(players, pokedex_dict, 2)
````

````python
# possible for function to work without pokedex_dict
# because pokemons' base stats are all listed out in the player_pokemon keys in players dictionary
def player_power(players_dict, player_id):
    player_power = 0

    for pokemon in players_dict[player_id]['player_pokemon']:
        player_power += sum(players_dict[player_id]['player_pokemon'][pokemon])
    
    print("{}'s power is {}".format(players_dict[player_id]['player_name'],player_power))
    return player_power

player_power(players, 1)
````

Your end result is effectively a matrix. Each list in the outer list is a row, and the *j*th elements of list together form the *j*th column, which represents a data attribute. The first three lists in your pokedex list should look like this:

````python
split_line = raw_pd.splitlines()
split_line
````

````python
removed_quotes = []
for row in split_line:
    removed_quotes.append(row.replace('"',''))
    
removed_quotes
````

````python
#seems like there are no empty rows
for row in removed_quotes:
    if ',,' in row:
        print(row)
    elif ', ,' in row:
        print(row)
````

````python
# creating a step to impute blank values should we ever upload another file with blank values
impute_blanks = []

for row in removed_quotes:
    if ', ,' in row:
        impute_blanks.append(row.replace(', ,',', -1,'))
    elif ',,' in row:
        impute_blanks.append(row.replace(',,',', -1,'))
    else:
        impute_blanks.append(row)
                        
impute_blanks                    
````

````python
split_rows = []
for row in impute_blanks:
    split_rows.append(row.split(','))
pprint(split_rows)
````

````python
final_pokedex = []

for row in split_rows:
    final_rows = []
    # changing all numeric items to floats in a row
    for string in row:
        if string.isnumeric():
            final_rows.append(float(string))
        else:
            final_rows.append(string)
    final_pokedex.append(final_rows)
    final_rows = []
````

## 9. Write a function to generate the full pokedex

---

Write a function that recreates the pokedex you made before, but with the data read in from the full pokemon file. The `PokedexNumber` should be used as the `pokemon_id` key values for the dictionary of pokemon.

Your function should:

1. Take the parsed pokedex information you created above as an argument.
2. Return a dictionary in the same format as your original pokedex you created before containing the information from the parsed full pokedex file.

To test the function, print out the pokemon with id = 100.

#### Note: 
- The solution below is what I understand the question to be asking for: using PokedexNumber from the csv as the pokemon_id
- But there are duplicate PokedexNumbers for related pokemon (e.g. Venusaur and VenusaurMega)
- The resulting dictionary overwrites values with the same key with the latest value. E.g. Charizard with PokedexNumber 6 is replaced with Charizard Y in this dictionary
- This makes less sense to me, unless the intended behaviour is for the dictionary to hold the most evolved Pokemons

````python
final_pokedex
````

````python
def recreate_pokedex(pokedex_parsed):
    
    # storing the pokedex_id separately to use as the outer dictionary key
    pokedex_id = [int(row[0]) for row in pokedex_parsed[1:]]
    
    # [1:] to separate "PokedexNumber" header from attribute, and pokedex numbers from data
    # this is for the inner, nested dictionary
    attribute = pokedex_parsed[0][1:]
    data = [(row[1:]) for row in pokedex_parsed[1:]]
    
    return {k:dict(zip(attribute,data[i])) for i,k in enumerate(pokedex_id)}

recreate_pokedex(final_pokedex)[100]
````

#### Note:
- Made another function that will output a dictionary that displays all the pokemons (799) but the pokemon id does not follow the PokedexNumbers in the csv file

````python
def recreate_pokedex_2(pokedex_parsed):
    attribute = pokedex_parsed[0][1:]
    data = [(row[1:]) for row in pokedex_parsed[1:]]
    pokedex_id = [int(row[0]) for row in pokedex_parsed[1:]]
    
    # this takes the length of the pokedex_id, rather than the pokedex number itself
    # hence bypassing the duplicated pokedex numbers
    return {k:dict(zip(attribute,data[k-1])) for k in range(1, len(pokedex_id))}

recreate_pokedex_2(final_pokedex)
````

````python
def filter_pokedex(pokedex, filter_options=None):
    
    # set default as False, only if it proves otherwise do we append to filtered list
    meets_criteria = False 
    filtered_pokemons = []
    
    all_filter_attributes = list(filter_options.keys())
    
    # looping through all the pokemon
    for i in range(1, len(pokedex)+1):
        # looping through all attributes for each pokemon
        for filter_attribute in all_filter_attributes:
            if str(filter_options[filter_attribute]).isnumeric():
                if pokedex[i][filter_attribute] >= filter_options[filter_attribute]:
                    meets_criteria = True
                else:
                    meets_criteria = False
                    # break at this loop because if at least 1 criteria if False, the pokemon shouldn't be filtered
                    break

            elif filter_options[filter_attribute].isalpha():
                if pokedex[i][filter_attribute] == filter_options[filter_attribute]:
                    meets_criteria = True 
                else:
                    meets_criteria = False
                    break
        if meets_criteria:
            filtered_pokemons.append(i)
    
    # loop through filtered_pokemons list to return corresponding dictionary stats for each pokedex number
    return [pokedex[num] for num in filtered_pokemons]

filter_options = {
    'Attack': 30,
    'Defense': 50,
    'SpecialAttack': 70,
    'Total': 550,
    'Type': 'Electric'
}

filter_pokedex(pokedex,filter_options=filter_options)
````

## 10. Calibrate the frequency of Pokemon

The design team wants you to make the powerful Pokemon rare, and the weaklings more common. How would you set the probability $p_i$ of finding Pokemon *i* each time a player visits a gym?

Write a function that takes in a Pokedex number and returns a value $p_i$ for that character.

Hint: there are many ways you could do this. What do _you_ think makes sense? Start with simplifying assumptions: for example, you could assume that the probabilities of encountering any two Pokemon on one visit to a gym are independent of each other.

````python
def occurrence_prob(pokedex_number):
    
    pokemon_score = pokedex[pokedex_number]['Total']
    
    if pokemon_score < 200:
        probability = (1/(0.65*pokemon_score))*100
    elif 200 <= pokemon_score < 300:
        probability = (1/(0.73*pokemon_score))*100
    elif 300 <= pokemon_score < 400:
        probability = (1/(0.7*pokemon_score))*100
    elif 400 <= pokemon_score < 500:
        probability = (1/(0.7*pokemon_score))*100
    elif 500 <= pokemon_score < 650:
        probability = (1/(1*pokemon_score))*100
    elif 650 <= pokemon_score < 750:
        probability = (1/(1.25*pokemon_score))*100
    else: 
        probability = (1/(3*pokemon_score))*100
    
    pokemon_name = pokedex[pokedex_number]['Name']
    
    # uncomment to check probability for each pokemon:
    #print('Probability of finding {} is {:.2f}'.format(pokemon_name,probability))
    
    return probability
````

````python
def occurrence_dist():
    return [occurrence_prob(pokemon_id) for pokemon_id in pokedex]
````
